```python
%matplotlib inline
import xml.etree.ElementTree
from pathlib import Path
from PIL import Image
import glob
import numpy as np
from matplotlib.pyplot import imshow
from pylab import rcParams
import math
rcParams['figure.figsize'] =  15,15



data_folder =  Path("../DATA")
rotated_images_folder = data_folder / 'rotated_images'
raw_images_folder = data_folder / 'tisky_17_stoleti_komplet'
doc = xml.etree.ElementTree.parse('../annotation_data/rotation/Old_script_detection5.xml').getroot()
for elem in doc.findall('image'):
    print(elem.get('name'))
    image_name = elem.get('name')
    # get name of the folder
    name_parts = image_name.split("_")
    folder_name_substr = f'{name_parts[0]}_{name_parts[1]}_{name_parts[2]}_{name_parts[3]}'
    collection_folder = glob.glob(str(raw_images_folder / folder_name_substr ) + "*")[0]
    # load the image
    image_path = Path(collection_folder) / image_name
    image = Image.open(image_path)
    # compute the rotation if the element has lines
    if elem.getchildren()[0].tag == 'points':
        # save the image as it is
        image.save(rotated_images_folder / image_name)
        continue
    else:
        angles = []
        for line in elem.getchildren():
            if line.tag != 'polyline':
                raise Exception(f'The label was not polyline. Instead {line.tag}.Image: {image_name}')
            points_strs = line.get('points').split(';')
            if len(points_strs) != 2:
                raise Exception(f'The line had more than two vertices ({len(points_strs)}).Image: {image_name}')
            x1_str,y1_str,x2_str,y2_str = points_strs[0].split(',')+points_strs[1].split(',')
            drawn_l2r = (float(x1_str)<float(x2_str))
            print(drawn_l2r)
            v1,v2 = (float(x2_str)-float(x1_str)),(float(y1_str)-float(y2_str))
            angle = math.atan2(v2, v1) * 180 / math.pi
            angles.append(angle)
        average_angle = np.mean(angles)
        print(average_angle)
        # rotate the image
        rotated_image = image.rotate(-average_angle)
        # save the rotated image
        rotated_image.save(rotated_images_folder / image_name)
```
